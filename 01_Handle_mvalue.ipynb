### ->When deleting rows/columns is OK vs risky (dropna, filtering, etc.)

**Acceptable to delete when:**
- Missing values are **few** (e.g., <510%)
- Data is **missing completely at random (MCAR)**
- The column/row is **non-critical** for analysis or modeling
- Duplicates or obvious data-entry errors
- During **EDA** or quick prototyping
- Removing rows with **impossible values** (e.g., age < 0)

**Risky to delete when:**
- Missing values are **systematic** (not random)
- Deletion removes a **large portion** of data
- The column is **important** for prediction or interpretation
- It causes **class imbalance** (especially targets)
- Working with **small datasets**
- Data may contain **rare but valid cases**

**Rule of thumb:**
> Drop data only when it does **less harm than distortion**; otherwise, prefer **imputation or flagging missingness**.
